1. the set of all strings with two consecutive repeated words (e.g., "Humbert Humbert" and "the the" but not "the bug" or "the big bug");
2. all strings that start at the beginning of the line with an integer and that end at the end of the line with a word;
3. all strings that have both the word _grotto_ and the word _raven_ in them (but not, e.g., words like _grottos_ that merely contain the word _grotto_);
4. write a pattern that places the first word of an English sentence in a register. Deal with punctuation.

```python
def inreset(compiled:re.Match, text:str) -> bool:
    "Tests to see if text is within the regex set defined by compiled."
    check = compiled.match(text)
    if check is None:
        return False
    return check.group()==text

# answers
ans_2_1_1 = re.compile('[A-Z|a-z]+')
ans_2_1_2 = re.compile('[a-z]*b')
ans_2_1_3 = re.compile('((bab)|b)+')

ans_2_2_1 = re.compile(r'(?P<wrd>\w+)[,\.!\?]?\s*(?P=wrd)')
ans_2_2_2 = re.compile(r'^[0-9].*\s\w+', flags=re.MULTILINE|re.DOTALL)
two_two_three = r'.*\s*(grotto)[,\.]?\s*.*\s*(raven)[,\.]?\s*.*|'
two_two_three += r'.*\s*(raven)[,\.]?\s*.*\s*(grotto)[,\.]?\s*.*'
ans_2_2_3 = re.compile(two_two_three, flags=re.DOTALL)
ans_2_2_4 = re.compile(r'[!?.]\s*')

answers = {'2.1.1': ans_2_1_1,
           '2.1.2': ans_2_1_2,
           '2.1.3': ans_2_1_3,
           '2.2.1': ans_2_2_1,
           '2.2.2': ans_2_2_2,
           '2.2.3': ans_2_2_3}
```

**2.6** Now implement a minimum edit distance algorithm and use your hand-computed results to check your code.

```python
def medist(source:str, target:str, ins_cost:int=1, del_cost:int=1,
           subst_cost:int=1, show_grid:bool=False) -> int:
    "Calculate the minimum edit distance between two strings."
    grid = pd.DataFrame(index=['#', *list(source)],
                        columns=['#', *list(target)])
    # First column
    for i in range(len(grid.index)):
        grid.iloc[i,0] = i
    # First row
    for i in range(len(grid.columns)):
        grid.iloc[0,i] = i
    # The rest
    for i,src in enumerate(grid.index[1:], 1):
        for j,tar in enumerate(grid.columns[1:], 1):
            left = grid.iloc[i-1,j]
            dcost = left+del_cost
            up = grid.iloc[i,j-1]
            icost = up+del_cost
            diag = grid.iloc[i-1,j-1]
            if src==tar:
                scost = diag
            else:
                scost = diag+subst_cost
            grid.iloc[i,j] = min(dcost, icost, scost)
    final = grid.iloc[-1,-1]
    if show_grid:
        print(grid)
    return final
```

```python
print('The edit distance of "leda" to "deal" is', medist("leda", "deal"), end=".\n")
print('The edit distance of "drive" to "brief" is', medist("drive", "brief"), end=".\n")
print('The edit distance of "drive" to "divers" is', medist("leda", "deal"), end=".")
```

**2.7** Augment the minimum edit distance algorithm to output an alignment; you will need to store pointers and add a stage to compute the backtrace.

```python
def medist_align(source:str, target:str, ins_cost:int=1, del_cost:int=1,
                 subst_cost:int=1, show_grid:bool=False) -> list:
    """
    Calculate the minimum edit distance and alignment between two
    strings and return the corresponding alignment.
    """
    grid = pd.DataFrame(index=['#', *list(source)],
                        columns=['#', *list(target)])
    # Top left cell
    grid.iloc[0,0] = 0
    # First column
    for i in range(len(grid.index))[1:]:
        grid.iloc[i,0] = (i,(i-1,0))
    # First row
    for i in range(len(grid.columns))[1:]:
        grid.iloc[0,i] = (i,(0,i-1))
    # The rest
    for i,src in enumerate(grid.index[1:], 1):
        for j,tar in enumerate(grid.columns[1:], 1):
            neighbours = {}
            left = grid.iloc[i-1,j]
            dcost = left[0]+del_cost
            neighbours[(i-1,j)] = dcost
            up = grid.iloc[i,j-1]
            icost = up[0]+del_cost
            neighbours[(i,j-1)] = icost
            diag = grid.iloc[i-1,j-1]
            if src==tar:
                scost = diag if type(diag)==int else diag[0]
            else:
                scost = subst_cost
                scost += diag if type(diag)==int else diag[0]
            neighbours[(i-1,j-1)] = scost
            minimum = min(dcost, icost, scost)
            grid.iloc[i,j] = (minimum, next(k for k in neighbours.keys()
                                            if neighbours[k]==minimum))
    if show_grid:
        print(grid)
    # Create a backtrace.
    active = (len(grid.index)-1,len(grid.columns)-1) # final cell
    backtrace = [active]
    while active!=(0,0):
        active = grid.iloc[active][1]
        backtrace.append(active)
    chain = backtrace[::-1]
    # Form the alignment.
    alignment = []
    src_used = []
    tar_used = []
    for fst,snd in chain[1:]: # 0,0 doesn't tell us anything.
        # print(fst,snd)
        if not(fst in src_used or snd in tar_used):
            operation = ('' if grid.index[fst]==grid.columns[snd] 
                         else 'del' if grid.columns[snd]=='#'
                         else 'ins' if grid.index[fst]=='#'
                         else 'sub')
            alignment.append((grid.index[fst], grid.columns[snd], operation))
        elif fst in src_used:
            alignment.append(('#', grid.columns[snd], 'ins'))
        else: #snd in tar_used
            True
            alignment.append((grid.index[fst], '#', 'del'))
        src_used.append(fst)
        tar_used.append(snd)
    return alignment
```

**3.8** Write a program to compute unsmoothed unigrams and bigrams.

```python
def unigram_count(word:str, text:str) -> Fraction:
    word = word.lower()
    text_list = [w.strip(",\"'.!?") for w in text.split()]
    numerator = len([w for w in text_list if w.lower()==word])
    denominator = len(text_list)
    return Fraction(numerator,denominator)

def bigram_count(word2:str, word1:str, text:str) -> Fraction:
    word2, word1 = (w.lower() for w in [word2, word1])
    sentences = [f"<s> {sentence} </s>" for sentence in re.split(r'[\.!?]', text)]
    numerator = 0
    denominator = 0
    for sentence in sentences:
        for i in range(len(sentence.split())-1):
            j = i+1
            if sentence.split()[i].strip(",\"'").lower()==word1:
                denominator += 1
                if sentence.split()[j].strip(",\"'").lower()==word2:
                    numerator += 1
    return Fraction(numerator,denominator)
```

```python
sam_text = "I am Sam. Sam I am. I am Sam. I do not like green eggs and Sam."
```

```python
print(unigram_count('Sam', sam_text))
```

```python
print(bigram_count("Sam", "am", sam_text))
```

```python
# This is too slow in practice, because of the repeated looping.
# def grams(text:str, include_zeroes:bool=False) -> dict:
#     words = {w.strip(",\"'.!?") for w in text.split()}
#     output = {}
#     # unigrams
#     for word in words:
#         output[word.lower()] = unigram_count(word, text)
#     # bigrams
#     for word0 in words:
#         for word1 in words.difference({word0}):
#             output[f"{word0.lower()} {word1.lower()}"] = bigram_count(word1, word0, text)
#     if include_zeroes:
#         return output
#     return {k:v for k,v in output.items() if v>0}

def grams(text:str) -> dict:
    output = {'unigrams':{}, 'bigrams':{}}
    # unigrams
    tokens = [w.strip(";,\"'.!?").lower() for w in text.split()]
    types = set(tokens)
    for word in types:
        output['unigrams'][word] = Fraction(tokens.count(word),len(tokens))
    # bigrams
    sentences = [f"<s> {sentence} </s>" for sentence in re.split(r'[\.!?]', text)
                 if len(sentence)>0]
    for sentence in sentences:
        words = [w.strip(";,\"'").lower() for w in sentence.split()]
        for i in range(len(words)-1):
            bigram = " ".join(words[i:i+2])
            denominator = tokens.count(words[i]) if i>0 else len(sentences) #<s> at the start of each sentence
            if bigram not in output['bigrams'].keys():
                output['bigrams'][bigram] = Fraction(1,denominator)
            else:
                output['bigrams'][bigram] += Fraction(1,denominator)
    return output
```

```python
both_grams = grams(sam_text)
print('Unigrams\n--------')
for k,v in both_grams['unigrams'].items():
    print(f"{k}\t{v}")
print('\nBigrams\n-------')
for k,v in both_grams['bigrams'].items():
    print(f"{k}{"\t"*(1-int(len(k)/8)+1)}{v}")
```

```python
# I'll take this as an instruction to generate random sentences based on a bigram
def random_sentence(bigrams:dict) -> str:
    output = ["<s>"]
    finished = False
    while not finished:
        candidates = [(k,v) for k,v in bigrams.items() if k.split()[0]==output[-1]]
        selection = choices([c[0] for c in candidates],
                            weights=[c[1] for c in candidates])[0].split()[-1]
        output.append(selection)
        if output[-1]=='</s>':
            finished = True
    return " ".join(output[1:-1])+'.'
```

```python
random_sentence(grams(udhr)['bigrams'])
```

```python
random_sentence(grams(first_john)['bigrams'])
```

**3.11** Add an option to your program to compute the perplexity of a test set.

```python
# I'll write separate functions for unigram and bigram model perplexity.
def unigram_perplexity(test_set:str, unigrams:dict) -> float:
    tokens = [w.strip(";,\"'.!?").lower() for w in test_set.split()]
    num_words = len(tokens)
    # log_denom = 0
    # for token in tokens:
    #     try:
            # log_denom += np.log(unigrams[token])
        # except TypeError:
        #     print(token)
    log_denom = sum([np.log(float(unigrams[token])) for token in tokens])
    return (1/np.exp(log_denom))**(1/num_words)
```

```python
# The probabilities come out underflowingly small for the longer texts.
unigram_perplexity(sam_text, grams(sam_text)['unigrams'])
```

```python
def bigram_perplexity(test_set:str, bigrams:dict) -> float:
    sentences = [f"<s> {sentence} </s>" for sentence in re.split(r'[\.!?]', test_set)
                 if len(sentence)>0]
    pairs = []
    for sentence in sentences:
        words = [w.strip(";,\"'").lower() for w in sentence.split()]
        for i in range(len(words)-1):
            pairs.append(" ".join(words[i:i+2]))
    log_denom = sum([np.log(float(bigrams[pair])) for pair in pairs])
    num_bigrams = len(pairs)
    return (1/np.exp(log_denom))**(1/num_bigrams)
```

```python
bigram_perplexity(sam_text, grams(sam_text)['bigrams'])
```

**3.12** You are given a training set of 100 numbers that consists of 91 zeros and 1 each of the other digits 1-9. Now we see the following test set: 0 0 0 0 0 3 0 0 0 0. What is the unigram perplexity?

- By hand: \begin{align*}
\textrm{perplexity}(W)&=\sqrt[N]{\prod_{i=1}^{N}\frac{1}{P(w_i)}}\\
&=\sqrt[10]{\frac{1}{P(0)^9\times P(3)}}\\
&=\sqrt[10]{\frac{1}{0.91^9\times 0.01}}\\

What class will Naive bayes assign to the sentence “I always like foreign films.”?

- $$\hat{c}=\textrm{argmax}_{c\in C}P(c|d)=\textrm{argmax}_{c\in C}P(d|c)P(c)$$
- Since the prior probabilities are equal, we can ignore $P(c)$ and move onto the probabilites for the individual words.
- $$\hat{P}(\textrm{I always like foreign films}|\textbf{pos})=0.09\times 0.07\times 0.29\times 0.04\times 0.08$$

```python
print("=",0.16*0.06*0.06*0.15*0.11)
```

- Naive Bayes will assign the sentence "I always like foreign films" to the class **negative**.

**4.3** Train two models, multinomial naive Bayes and binarized naive Bayes, both with add-1 smoothing, on the following document counts for key sentiment words, with positive or negative class assigned as noted.

| doc | “good” | “poor” | “great” | (class) |
| - | - | - | - | - |
| d1. | 3 | 0 | 3 | pos |
| d2. | 0 | 1 | 2 | pos |
| d3. | 1 | 3 | 0 | neg |
| d4. | 1 | 5 | 2 | neg |
| d5. | 0 | 2 | 0 | neg |

Use both naive Bayes models to assign a class (pos or neg) to this sentence:

> A good, good plot and great characters, but poor acting.

Recall from page 61 that with naive Bayes text classification, we simply ignore (throw out) any word that never occurred in the training document. (We
don’t throw out words that appear in some classes but not others; that’s what add-one smoothing is for.) Do the two models agree or disagree?

- I don't see how I'm supposed to do this without the word counts for each of the documents. For the purposes of this exercise, I will assume that each document has length 10.

```python
# Store the results from the above table in a dictionary.
multinom = {1: {'pos': True,
                'counts': {'good': 3,
                           'poor': 0,
                           'great': 3}},
            2: {'pos': True,
                'counts': {'good': 0,
                           'poor': 1,
                           'great': 2}},
            3: {'pos': False,
                'counts': {'good': 1,
                           'poor': 3,
                           'great': 0}},
            4: {'pos': False,
                'counts': {'good': 1,
                           'poor': 5,
                           'great': 2}},
            5: {'pos': False,
                'counts': {'good': 0,
                           'poor': 2,
                           'great': 0}}}
# For the binarized case, we clip the counts at 1.
binar = {1: {'pos': True,
             'counts': {'good': 1,
                        'poor': 0,
                        'great': 1}},
         2: {'pos': True,
             'counts': {'good': 0,
                        'poor': 1,
                        'great': 1}},
         3: {'pos': False,
             'counts': {'good': 1,
                        'poor': 1,
                        'great': 0}},
         4: {'pos': False,
             'counts': {'good': 1,
                        'poor': 1,
                        'great': 1}},
         5: {'pos': False,
             'counts': {'good': 0,
                        'poor': 1,
                        'great': 0}}}
# Return P(c|d) for each c.
def posterior(text:str, counts_dict:dict) -> dict:
    tokens = [w.strip(",.!\"'?").lower() for w in text.split(" ")]
    poss = {k:v['counts'] for k,v in counts_dict.items() if v['pos']}
    negs = {k:v['counts'] for k,v in counts_dict.items() if not v['pos']}
    output = {}
    for i,cls in enumerate([poss,negs]):
        # Start with the prior probability.
        res = len(cls.keys())/(len(poss.keys())+len(negs.keys()))
        # Multiply by each likelihood in turn.
        for token in tokens:
            for d in cls.values():
                if token in d.keys():
                    numerator = d[token]+1
                    denominator = 11 #10 tokens in each sentence + 1
                    res *= Fraction(numerator,denominator)
        if i==0:
            output['pos'] = res
        else:
            output['neg'] = res
    return output
```

```python
test_s = 'A good, good plot and great characters, but poor acting.'
print(f"""Multinomial naive Bayes
-----------------------
Positive probability:\t{posterior(test_s, multinom)['pos']}
Negative probability:\t{posterior(test_s, multinom)['neg']}

Binarized naive Bayes
---------------------
Positive probability:\t{posterior(test_s, binar)['pos']}
Negative probability:\t{posterior(test_s, binar)['neg']}""")
```

- The two models agree on the classification of the sentence.
