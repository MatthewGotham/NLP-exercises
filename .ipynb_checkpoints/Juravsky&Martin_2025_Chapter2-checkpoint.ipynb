**2.1** Write regular expressions for the following languages.

1. the set of all alphabetic strings;
2. the set of all lower case alphabetic strings ending in a _b_;

**2.2** Write regular expressions for the following languages. By "word", we mean an alphabetic string separated from other words by whitespace, any relevant punctuation, line breaks, and so forth.

1. the set of all strings with two consecutive repeated words (e.g., "Humbert Humbert" and "the the" but not "the bug" or "the big bug");
2. all strings that start at the beginning of the line with an integer and that end at the end of the line with a word;
3. all strings that have both the word _grotto_ and the word _raven_ in them (but not, e.g., words like _grottos_ that merely contain the word _grotto_);
4. write a pattern that places the first word of an English sentence in a register. Deal with punctuation.

```python
def inreset(compiled:re.Match, text:str) -> bool:
    "Tests to see if text is within the regex set defined by compiled."
    check = compiled.match(text)
    if check is None:
        return False
    return check.group()==text

# answers
ans_2_1_1 = re.compile('[A-Z|a-z]+')
ans_2_1_2 = re.compile('[a-z]*b')
ans_2_1_3 = re.compile('((bab)|b)+')

ans_2_2_1 = re.compile(r'(?P<wrd>\w+)[,\.!\?]?\s*(?P=wrd)')
ans_2_2_2 = re.compile(r'^[0-9].*\s\w+', flags=re.MULTILINE|re.DOTALL)
two_two_three = r'.*\s*(grotto)[,\.]?\s*.*\s*(raven)[,\.]?\s*.*|'
two_two_three += r'.*\s*(raven)[,\.]?\s*.*\s*(grotto)[,\.]?\s*.*'
ans_2_2_3 = re.compile(two_two_three, flags=re.DOTALL)
ans_2_2_4 = re.compile(r'[!?.]\s*')

answers = {'2.1.1': ans_2_1_1,
           '2.1.2': ans_2_1_2,
           '2.1.3': ans_2_1_3,
           '2.2.1': ans_2_2_1,
           '2.2.2': ans_2_2_2,
           '2.2.3': ans_2_2_3}
```

```python
test_set = ['Windows','','asdf4s','asDf','Blub','blub','blub1','babb',
            'rabb','b','Humbert Humbert','the the','the bug','the big bug',
            'word,word', 'Word. Word', 'word\n word', '3\nis the magic number',
            'wibble, grotto and raven', 'raven grotto', 'raven grottos']
```

```python
# check
for anstring,answer in answers.items():
    print(f"Answer {anstring}: {answer}. Checks:\n")
    for i,test in enumerate(test_set):
        print(f"'{test}' in {answer}?",'\t',inreset(answer, test))
        if i+1==len(test_set):
            print('-----\n')
```

**2.3** Implement an ELIZA-like program, using substitutions such as those described on page 12. You might want to choose a different domain than a Rogerian psychologist, although keep in mind that you would need a domain in which your program can legitimately engage in a lot of simple repetition.

```python
# answer
while True:
    patient = input('> ')
    if patient.upper()=="STOP":
        break
    if 'all' in patient.lower():
        print("IN WHAT WAY?")
    elif 'always' in patient.lower():
        print("CAN YOU GIVE A SPECIFIC EXAMPLE?")
    else:
        look = r"(I'M FEELING|I'M|I AM FEELING|I AM|I FEEL) (?P<fl>\w*)"
        feeling = re.search(look, patient.upper())
        if feeling is not None:
            emotion = feeling.group('fl')
            if emotion in ['DEPRESSED','SAD','LOW']:
                print(f"I AM SORRY TO HEAR THAT YOU ARE {emotion}.")
            elif emotion in ['HAPPY','JOYFUL','GLAD','OVERJOYED','ELATED']:
                print(f"I AM GLAD TO HEAR THAT YOU ARE {emotion}.")
            else:
                print(f"YOU ARE {emotion}.")
        else:
            amended = patient.upper()
            for pair in [(r"(I'M)|(I AM)", "YOU ARE"), (r"(I )|(ME )", "YOU "),
                              ("(MY )", "YOUR "), ("MINE ", "YOURS ")]:
                amended = re.sub(pair[0], pair[1], amended)
            print(amended)
```

**2.6** Now implement a minimum edit distance algorithm and use your hand-computed results to check your code.

```python
def medist(source:str, target:str, ins_cost:int=1, del_cost:int=1,
           subst_cost:int=1, show_grid:bool=False) -> int:
    "Calculate the minimum edit distance between two strings."
    grid = pd.DataFrame(index=['#', *list(source)],
                        columns=['#', *list(target)])
    # First column
    for i in range(len(grid.index)):
        grid.iloc[i,0] = i
    # First row
    for i in range(len(grid.columns)):
        grid.iloc[0,i] = i
    # The rest
    for i,src in enumerate(grid.index[1:], 1):
        for j,tar in enumerate(grid.columns[1:], 1):
            left = grid.iloc[i-1,j]
            dcost = left+del_cost
            up = grid.iloc[i,j-1]
            icost = up+del_cost
            diag = grid.iloc[i-1,j-1]
            if src==tar:
                scost = diag
            else:
                scost = diag+subst_cost
            grid.iloc[i,j] = min(dcost, icost, scost)
    final = grid.iloc[-1,-1]
    if show_grid:
        print(grid)
    return final
```

```python
print('The edit distance of "leda" to "deal" is', medist("leda", "deal"), end=".\n")
print('The edit distance of "drive" to "brief" is', medist("drive", "brief"), end=".\n")
print('The edit distance of "drive" to "divers" is', medist("leda", "deal"), end=".")
```

```python
medist("leda", "deal", show_grid=True)
```

**2.7** Augment the minimum edit distance algorithm to output an alignment; you will need to store pointers and add a stage to compute the backtrace.

```python
def medist_align(source:str, target:str, ins_cost:int=1, del_cost:int=1,
                 subst_cost:int=1, show_grid:bool=False) -> list:
    """
    Calculate the minimum edit distance and alignment between two
    strings and return the corresponding alignment.
    """
    grid = pd.DataFrame(index=['#', *list(source)],
                        columns=['#', *list(target)])
    # Top left cell
    grid.iloc[0,0] = 0
    # First column
    for i in range(len(grid.index))[1:]:
        grid.iloc[i,0] = (i,(i-1,0))
    # First row
    for i in range(len(grid.columns))[1:]:
        grid.iloc[0,i] = (i,(0,i-1))
    # The rest
    for i,src in enumerate(grid.index[1:], 1):
        for j,tar in enumerate(grid.columns[1:], 1):
            neighbours = {}
            left = grid.iloc[i-1,j]
            dcost = left[0]+del_cost
            neighbours[(i-1,j)] = dcost
            up = grid.iloc[i,j-1]
            icost = up[0]+del_cost
            neighbours[(i,j-1)] = icost
            diag = grid.iloc[i-1,j-1]
            if src==tar:
                scost = diag if type(diag)==int else diag[0]
            else:
                scost = subst_cost
                scost += diag if type(diag)==int else diag[0]
            neighbours[(i-1,j-1)] = scost
            minimum = min(dcost, icost, scost)
            grid.iloc[i,j] = (minimum, next(k for k in neighbours.keys()
                                            if neighbours[k]==minimum))
    if show_grid:
        print(grid)
    # Create a backtrace.
    active = (len(grid.index)-1,len(grid.columns)-1) # final cell
    backtrace = [active]
    while active!=(0,0):
        active = grid.iloc[active][1]
        backtrace.append(active)
    chain = backtrace[::-1]
    # Form the alignment.
    alignment = []
    src_used = []
    tar_used = []
    for fst,snd in chain[1:]: # 0,0 doesn't tell us anything.
        # print(fst,snd)
        if not(fst in src_used or snd in tar_used):
            operation = ('' if grid.index[fst]==grid.columns[snd] 
                         else 'del' if grid.columns[snd]=='#'
                         else 'ins' if grid.index[fst]=='#'
                         else 'sub')
            alignment.append((grid.index[fst], grid.columns[snd], operation))
        elif fst in src_used:
            alignment.append(('#', grid.columns[snd], 'ins'))
        else: #snd in tar_used
            True
            alignment.append((grid.index[fst], '#', 'del'))
        src_used.append(fst)
        tar_used.append(snd)
    return alignment
```
